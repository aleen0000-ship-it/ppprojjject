```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#load data
#we read the csv file into a dataframe and look at the first few rows.
```

```python
#load data

url_train= "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url_train)

#show first 5 rows to see the data
df.head()
```

```python
## Step 1: Fix Data Types

#`Pclass` is numeric but represents categories, so we convert it to `category`.
#Embarked` is also categorical (port codes), convert to `category`.
```

```python
def fix_data_types(df):
    df = df.copy()
    # Pclass is numeric but actually a category
    df["Pclass"] = df["Pclass"].astype("category")
    # Embarked is categorical (port codes)
    df["Embarked"] = df["Embarked"].astype("category")
    return df
```

```python
## Step 2: Handle Missing Values

#- `Age`: fill missing with median (robust to outliers).
#- `Embarked`: fill missing with mode (most common value).
#- `Fare`: fill missing with median.
#- `Cabin`: too many missing values, drop the column.
```

```python
def handle_missing_values(df):
    df = df.copy()
    df["Age"] = df["Age"].fillna(df["Age"].median())  # fill missing Age
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])  # fill missing Embarked
    df["Fare"] = df["Fare"].fillna(df["Fare"].median())  # fill missing Fare
    df = df.drop(columns=["Cabin"], errors="ignore")  # drop Cabin if exists, ignore if not
    return df
```

```python
## Step 3: Remove Duplicates

#Check for and remove any duplicate rows.
```

```python
def remove_duplicates(df):
    df = df.copy()
    # drop any duplicate rows if they exist
    df = df.drop_duplicates()
    return df
```

```python
## Step 4: Cap Outliers

#- Cap `Fare` at the 99th percentile to avoid extreme values affecting analysis.
```

```python
def cap_outliers(df):
    df = df.copy()
    # cap Fare at the 99th percentile to limit extreme values
    upper_fare = df["Fare"].quantile(0.99)
    df["Fare"] = np.where(df["Fare"] > upper_fare, upper_fare, df["Fare"])
    return df
```

```python
## Step 5: Ensure Positive Fare

#- Keep only rows where Fare is zero or positive.
```

```python
# -----------------------
def ensure_positive_fare(df):
    df = df.copy()
    # keep only rows where Fare is zero or positive
    df = df[df["Fare"] >= 0]
    return df
```

```python
## Step 6: Combine All Steps

#We create a main function `clean_data()` that runs all the cleaning steps in order.
```

```python
def clean_data(df):
    df = df.copy()
    # fix the data types first
    df = fix_data_types(df)
    # fill missing values and drop Cabin
    df = handle_missing_values(df)
    # remove duplicate rows if any
    df = remove_duplicates(df)
    # cap extreme Fare values at 99th percentile
    df = cap_outliers(df)
    # make sure Fare is not negative
    df = ensure_positive_fare(df)
    return df
```

```python
## Step 7: Apply Cleaning and Check Results
```

```python
df = clean_data(df)

# final checks after cleaning
assert df["Age"].isnull().sum() == 0, "There are missing values in Age!"
assert df["Fare"].isnull().sum() == 0, "There are missing values in Fare!"
assert (df["Fare"] >= 0).all(), "Some Fare values are negative!"
assert df.shape[1] == 11, f"Expected 11 columns, got {df.shape[1]}"

# Show final shape and first 5 rows
print("Final shape:", df.shape)
df.head()
```
